```python
import tensorflow as tf
from tensorflow import keras 
tf.__version__
from keras import Sequential,datasets
from tensorflow.python.keras.layers import Dense, Conv2D, Flatten, Dropout, MaxPooling2D
image_data_generator = tf.keras.preprocessing.image.ImageDataGenerator()
import numpy as np
import os
```

```python
os.version['TF_XLA_FLAGS']  ='--tf_xla_enable_xla_devices'
```

### Load DataSet

```python
(x_train, y_train),(x_test,y_test)=datasets.cifar10.load_data()
```

```python
x_train.shape
```

```python
x_test.shape
```

### Reshape

```python
y_train=y_train.reshape(-1, )
y_train
```

```python
y_classes = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]
```

```python
len(y_classes)
```

### Observe Dataset

```python
def showImage(x,y,index):
    plt.figure(figsize=(15,2))
    plt.imshow(x[index])
    plt.xlabel(y_classes[y[index]])
```

```python
showImage(x_train, y_train,4)
```

### Normalize Dataset

```python
x_train[0]
```

```python
x_train=x_train/255
x_test=x_test/255
x_train[0]
```

### Build Model

```python
model=Sequential()
#Convolution Layer
model.add(Conv2D(filters=32, kernel_size=(3,3),activation="relu",input_shape=(32,32,32)))
#Adding max pooling(2x2)
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(filters=64,kernel_size=(4,4),activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
#Flatten Layer
model.add(Flatten())
model.add(Dense(units=34, activation="relu"))
model.add(Dense(units=10, activation="softmax"))
```

### Compile Model

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

### Train Model

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=5)
```

### Make Predictions

```python
y_predictions = model.predict(x_test)
```

```python
y_predictions[9]
```

```python
y_predictions=[np.argmax(arr) for arr in y_predictions]
```
